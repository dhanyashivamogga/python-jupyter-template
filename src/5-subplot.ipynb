### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

bar_x = ['A', 'B', 'C', 'D']
bar_y = [5, 7, 3, 4]

scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)

pie_labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
pie_sizes = [25, 35, 20, 20]

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Chart')

# Bar chart
axs[0, 1].bar(bar_x, bar_y, color='green')
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='red')
axs[1, 0].set_title('Scatter Plot')

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
